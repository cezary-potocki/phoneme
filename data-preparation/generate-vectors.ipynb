```python
import numpy as np 
import pandas as pd 
import pyphi
import requests
```

```python
def call_embedding_ws(names):
#    url = 'http://54.36.53.127:8009/embedding'
    url = 'http://127.0.0.1:8009/embedding'
    headers = {"content-type": "application/json"}
    response = requests.post(url, json=names)
    
    arr = pyphi.jsonify.loads(response.text)
    x = np.array(arr)
    x = x.astype(dtype=np.float32)
    
    return x
```

```python
%%time
eng_names = pd.read_csv('data/ENG_FULL_NAMES.csv', sep='|', index_col='id')
i = 0
ws_params = []

file = open("data/ENG_VECTORS.csv","w") 

for index, row in eng_names.iterrows():
    ws_params.append(row['eng'].upper())
    i = i + 1
    if i % 512 == 0:
        ws_result = call_embedding_ws(ws_params)
        for vect in ws_result:
            v_str = np.array2string(vect, separator=',', max_line_width=10000).strip("[]").replace(" ", "")
            #eng_vectors = np.append(eng_vectors, v_str)
            file.write("{v}\n".format(v=v_str))
        ws_params = []
        print(i)
        
if len(ws_params) > 0:
    ws_result = call_embedding_ws(ws_params)
    for vect in ws_result:
        v_str = np.array2string(vect, separator=',', max_line_width=10000).strip("[]").replace(" ", "")
        #eng_vectors = np.append(eng_vectors, v_str)
        file.write("{v}\n".format(v=v_str))
        
file.close()
```
